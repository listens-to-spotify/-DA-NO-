```python
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
warnings.filterwarnings('ignore')
```

```python
df_orig = pd.read_csv('processed-hh-ru.csv')
```

```python
df = df_orig.copy()
```

```python
df.shape
```

```python
# Создайте пустой DataFrame для хранения результатов
result_list = []

# Группируйте данные по региону и выполните t-тест для каждой группы
for region, group in df.groupby('vacancy_region'):
    successful = group[group['successful_topic'] == 1]
    unsuccessful = group[group['successful_topic'] == 0]
    
    t_stat, p_value = ttest_ind(successful['work_experience_months'], unsuccessful['work_experience_months'])
    
    result_list.append({
        'Region': region,
        't-statistic': t_stat,
        'p-value': p_value
    })

# Создайте DataFrame из списка результатов
result_df = pd.DataFrame(result_list)

# Отсортируйте результаты по значению p-value
result_df = result_df.sort_values(by='p-value')

# Выведите результаты
result_df
```

```python
result_df['p-value'] = result_df['p-value'].apply(lambda x: round(x, 3))
```
